```python
df_1 = unicorn.loc[:, ['Valuation ($B)']].sort_values(by='Valuation ($B)')
fig = px.box(df_1,y="Valuation ($B)",color_discrete_sequence=["mediumturquoise"],template='plotly_dark',height=700,width=1500)
fig.update_traces(quartilemethod="exclusive", jitter=0.8)

fig.show()
```
